# Scrap Population Data

```python
## load the dependecies
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
## create a driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
## set the url for scraping
target_url = 'https://www.scrapethissite.com/pages/simple/'
driver.get("https://www.scrapethissite.com/pages/simple/")
print(driver.title)
```

```python
## scrap the country data, its capital, its area, its population
country_list = driver.find_elements(By.XPATH,"//h3[@class='country-name']")
print(country_list)

capital_list = driver.find_elements(By.XPATH,"//span[@class='country-capital']")
print(capital_list)

population_list = driver.find_elements(By.XPATH,"//span[@class='country-population']")
print(population_list)

area_list = driver.find_elements(By.XPATH,"//span[@class='country-area']")
print(area_list)
```

```python
## create a country list
countries = []
for country in country_list:
    country_name = country.text
    countries.append(country_name)

print(countries)
```

```python
## create a population list
population = []
for pop in population_list:
    population_size = int(pop.text)
    population.append(population_size)

print(population)
```

```python
## create a capital list
capital = []
for cap in capital_list:
    capital_city = cap.text
    capital.append(capital_city)

print(capital)
```

```python
## create an area list
area = []
for pop in area_list:
    area_size = float(pop.text)
    area.append(area_size)

print(area)
```

```python
## get the population density
population_density = []

for i in range(len(population)):
    try :
        density = float(population[i])/float(area[i])
    except :
        density = 0
    population_density.append(density)

print(population_density)
```

```python
## create a pandas dataframe
dict = {
         'country':[],
         'capital':[],
         'population size':[],
         'area(in sq.km)':[],
         'population density(person/sq.km)':[]
       }

dict['country']=countries
dict['capital']=capital
dict['population size']=population
dict['area(in sq.km)']=area
dict['population density(person/sq.km)']=population_density
```

```python
## create and save the dataframe
df = pd.DataFrame(dict)
print(df.head())
```

```python
## save the data as csv and xlsx
df.to_csv('country_wise_data.csv',index=False)
print("CSV data stored successfully")

df.to_excel('country_wise_data.xlsx',index=False)
print("Excel data stored successfully")
```

```python
## visualisation 
plt.figure(figsize=(25, 25))
plt.bar(countries[50:100], height=population[50:100])
plt.xticks(rotation=90, ha="center")
plt.show()
```

```python
# Sort by population size
df_sorted = df.sort_values(by="population size", ascending=False)

# Top 15 + others
top10 = df_sorted.head(15)
others = pd.DataFrame({
    "country": ["Others"],
    "population size": [df_sorted["population size"].iloc[15:].sum()]
})

final_df = pd.concat([top10, others], ignore_index=True)

#  Pie chart (use correct column name)
plt.figure(figsize=(8,8))
plt.pie(
    final_df["population size"],
    labels=final_df["country"],
    autopct="%1.1f%%",
    startangle=140
)
plt.title("Population Distribution (Top 15 + Others)", fontsize=14)
plt.tight_layout()
plt.show()
```
